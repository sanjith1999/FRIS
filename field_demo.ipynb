### Sequential Run

```python
from libs.forward_lib.field_model import FieldModel
from tqdm import tqdm

nx, ny, nz = 256, 256, 256

for it in tqdm(range(0, 10), desc ="H2 Calculation: "):
    FieldModel.NA = it*.1+0.5
    FM = FieldModel(nx,ny, nz, device='cpu')
    FM.propagate_field()
    FM.save_object_space(it = it+1000)
    
```
