### Sequential Run

```python
nx, ny, nz = 200, 200, 200
for it in range(10):
    FieldModel.NA = it*.1+0.5
    FM = FieldModel(nx,ny, nz)
    FM.propagate_field()
    FM.save_object_space(it = it+1)
    
```
